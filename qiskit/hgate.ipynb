基于 https://github.com/Qiskit/qiskit-tutorials/blob/master/tutorials/circuits/2_plotting_data_in_qiskit.ipynb

```python
from qiskit import *
from qiskit.visualization import plot_histogram
# quantum circuit to make a Bell state 
bell = QuantumCircuit(1)
bell.h(0)
#bell.cx(0, 1)

meas = QuantumCircuit(1)
meas.measure_all()

# execute the quantum circuit 
backend = BasicAer.get_backend('qasm_simulator') # the device to run on
circ = bell.compose(meas)
result = backend.run(transpile(circ, backend), shots=1).result()
counts  = result.get_counts(circ)
print(counts)
```

```python
plot_histogram(counts)
```

```python
# Execute 2-qubit Bell state again
second_result = backend.run(transpile(circ, backend), shots=1).result()
second_counts  = second_result.get_counts(circ)

# Execute 2-qubit Bell state again
thrid_result = backend.run(transpile(circ, backend), shots=1).result()
thrid_counts  = thrid_result.get_counts(circ)

# Execute 2-qubit Bell state again
fourth_result = backend.run(transpile(circ, backend), shots=1).result()
fourth_counts  = fourth_result.get_counts(circ)
# Plot results with legend

# Execute 2-qubit Bell state again
fifth_result = backend.run(transpile(circ, backend), shots=1).result()
fifth_counts  = fifth_result.get_counts(circ)

# Execute 2-qubit Bell state again
sixth_result = backend.run(transpile(circ, backend), shots=1).result()
sixth_counts  = sixth_result.get_counts(circ)

# Execute 2-qubit Bell state again
seventh_result = backend.run(transpile(circ, backend), shots=1).result()
seventh_counts  = seventh_result.get_counts(circ)

legend = ['1st execution', '2nd execution','3rd execution','4th execution','5th execution','6th execution','7th execution']
plot_histogram([counts, second_counts, thrid_counts, fourth_counts,fifth_counts,sixth_counts,seventh_counts],
color=['red','orange','yellow','green','blue','midnightblue','black'],
 legend=legend)
```

如图所示 7次实验 依次以 红，橙，黄，绿，蓝，深蓝，黑 表示
可以看到 第1次和第5次结果为0

```python
# Plot results with legend
# Execute 2-qubit Bell state again
thrid_k_result = backend.run(transpile(circ, backend), shots=1000).result()
thrid_k_counts  = thrid_k_result.get_counts(circ)

# Execute 2-qubit Bell state again
fourth_k_result = backend.run(transpile(circ, backend), shots=1000).result()
fourth_k_counts  = fourth_k_result.get_counts(circ)
legend = ['3k execution', '4k execution']
plot_histogram([thrid_k_counts, fourth_k_counts], legend=legend)
```

在两轮，每轮1千次实验的情况下
第一轮出现了505次结果为0
第二轮出现了495次结果为0
